##### Проверка эффективности обогащения

###### формула расчета эффективности обогащения

```python
def get_recovery_au(row):
    # Зададим переменные для лучшего понимания и удобства
    ## C — доля золота в концентрате после флотации/очистки
    C = row['rougher.output.concentrate_au']
    ## F — доля золота в сырье/концентрате до флотации/очистки
    F = row['rougher.input.feed_au']    
    ## T — доля золота в отвальных хвостах после флотации/очистки
    T = row['rougher.output.tail_au']
    #Расчет
    try:
        recovery = ((C * (F - T)) / (F * (C - T))) * 100

        return recovery
    except:
        return 0
```

###### Проверка эффективности обогащения на обучающей выборке

```python
calculated_recovery_train = df_train.apply(get_recovery_au, axis=1)
mae_recovery_train = mean_absolute_error(df_train['rougher.output.recovery'], calculated_recovery_train)
print(f'MAE между нашими расчётами и истинными значением признака составляет {mae_recovery_train}')
```

```python
compared_recovery = pd.Series(calculated_recovery_train - df_train['rougher.output.recovery'])

print('Количество объектов полностью не равныъ формуле:', len(compared_recovery[compared_recovery != 0]))
print('Средняя абсолютная ошибка составляет:', mae_recovery_train)
```

## Модель

### Создадим функции оценки модели sMAPE

###### sMAPE

```python
def smape(target, pred):
    #target[np.isnan(target)] = 0
    #pred[np.isnan(pred)] = 0
    #t, p = np.array(target), np.array(pred)
    t, p = target, pred
    return np.mean(np.abs(t-p) / ((np.abs(t) + np.abs(p)) / 2) * 100)

smape_score = make_scorer(smape, greater_is_better=False)
```

###### Итоговое sMAPE

```python
def fin_smape(smape_r, smape_f):
    return (0.25 * smape_r + 0.75 * smape_f)

fin_smape_score = make_scorer(fin_smape, greater_is_better=False)
```

### Функция проверки модели

```python
def model_train(model_name, features, target):
    if model_name == 'dummy':
        parameters = {}
        dummy_random = RandomizedSearchCV(DummyRegressor(strategy='mean'), parameters, 
                                          scoring=smape_score, n_jobs=-1, random_state=RAND_STATE, cv=5)
        return dummy_random.fit(features, target)
    
    elif model_name == 'linear':
        parameters = {            
                     }
        lr_random = RandomizedSearchCV(LinearRegression(), parameters,
                                       scoring=smape_score, n_jobs=-1, random_state=RAND_STATE, cv=5)
        return lr_random.fit(features, target)
    
    elif model_name == 'tree':
        parameters = {'max_depth':list(np.arange(2, 20))
                     }
        tree_random = RandomizedSearchCV( DecisionTreeRegressor(random_state=RAND_STATE), parameters, n_iter=18, 
                                          scoring=smape_score, n_jobs=-1, random_state=RAND_STATE, cv=5)
        return tree_random.fit(features, target)
    
    elif model_name == 'forest':
        parameters = {'max_depth':list(np.arange(2, 20)),
                      'n_estimators':list(np.arange(4, 205, 20))
                     }
        forest_random = RandomizedSearchCV(RandomForestRegressor(random_state=RAND_STATE), parameters, n_iter=5, #20, 
                                           scoring=smape_score, n_jobs=-1, random_state=RAND_STATE, cv=5, verbose=2)
        
        return forest_random.fit(features, target)
    
    elif model_name == 'cat_boost':
        parameters = {'learning_rate':np.array([0.01, 0.1,0.25]), 
                      'n_estimators':list(np.arange(50, 200, 20)), 
                      'depth':list(np.arange(2, 20)) 
                     }
        cat_boost_random = RandomizedSearchCV(CatBoostRegressor(random_state=RAND_STATE, verbose=False), parameters, n_iter=5, #17,
                                              scoring=smape_score, n_jobs=-1, random_state=RAND_STATE, cv=5, verbose=2)
        return cat_boost_random.fit(features, target)
            
    pass
            
            
    
```
